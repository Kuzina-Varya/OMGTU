# Представлена задача регрессии, где необходимо спрогнозировать значение коэффициента «Гармония Бессмертия» — числового показателя, отражающего стабильность портала.

# ЦЦель — предсказать текущий уровень гармонии и определить, не достиг ли он критического значения, при котором портал становится нестабилен.

# Описание полей:
Вектор Мощи: Позиция магического механизма, управляющего интенсивностью перехода через портал.

Скорость перехода через портал: Скорость, с которой портал перемещает объект через пространство.(не может быть 0)

Приток Силы Потока: Поток магической силы, текущий через ядро портала, обеспечивая его стабильную работу.

Ритм магического ядра: Число оборотов магического ядра портала в минуту.

Поток Энергий: Число оборотов генератора эфира, преобразующего внешнюю энергию в магическую.

Сила Левого Потока: Мощность магического потока, текущего через левую сторону портала, поддерживая его баланс.

Сила Правого Потока: Мощность магического потока, текущего через правую сторону портала, обеспечивая равномерное распределение энергии.

Пламя Стихий: Температура магической энергии, исходящей из высокого магического источника портала, в градусах Цельсия.

Температура вдоха Истока: Температура воздуха, входящего в магический ускоритель портала, в градусах Цельсия.

Температура выдоха Истока: Температура воздуха, исходящего из магического ускорителя портала, в градусах Цельсия.

Приток давления Выдоха Истока: Давление магического потока на выходе из магического источника высокого давления.

Давление вдоха Истока: Давление воздуха, входящего в магический ускоритель.

Давление выдоха Истока: Давление воздуха, исходящего из магического ускорителя.

Древний Ветер: Давление древней магической энергии, покидающей портал в виде выхлопного потока.

Печать Чародея: Параметр управления магическим впрыском в сердце портала, выраженный в процентах.

Эмульсия Истока: Количество магического топлива, подпитывающего портал.(не может быть 0)

Дыхание Истока: Коэффициент, отображающий степень угасания магического источника.

Гармония Бессмертия — целевая переменная: Коэффициент, указывающий на состояние магического ядра портала и его стабильность, требующий магического восстановления в случае снижения.(не может быть 0)

Тип Вектора Мощи: Метка типа позиции магического механизма.

Номер пометки:Уникальный номер записи о работе портала в блокноте.(его будем отбрасывать, тк просто идентификатор)

# Зависимости:

(не добавляем как новые поля а то будет мультиколлениарноть)

1) Общая мощь=Сила левого потока+ Сила правого потока+Приток силы потока

2) Общая сила ядра=Ритм магического ядра* Приток силы потока

3) Общее давление на выходе= Приток давления выхода истока + Давление выода истока

4) Магическая производительность=Скорость перехода через портал/Эмульсия истока

5) Эффективность ядра= Общая сила ядра/ Эмульсия истока

6) Магическая мощьность=Эмульсия истока+ Общее давление на выходе+(Пламя стихий-Температура входа истока)

7) Степень износа магических источников= Дыхание истока/Гармония бессмертия(утечка)

8) Расхождение в стабильности магии = Дыхание истока- Гармония бессмертия(нельзя добавлять в обучение-утечка данных)

9) Баланс угасания= Расхождение в стабильности магии/ Скорость перехода через портал(нельзя добавлять в обучение-утечка)



```python
import pandas as pd

df = pd.read_csv("../data/portal_data.csv", sep="|", encoding="cp1251")
```

```python
#проанализируем

print(df.info())          # типы, пропуски
print(df.describe())      # базовая статистика чисел
print(df.nunique())       # уникальные значения (поймать ID/категориальные)
```

```python
#сделаем пропуски налом

import numpy as np

df.replace(["-", "Не определено"], np.nan, inplace=True)
```

```python
#преобразуем обжекты в флот (те что по смыслу должны быть флот)

num_cols = [
    "Скорость перехода через портал",
    "Температура выдоха Истока",
    "Приток давления Выдоха Истока",
    "Давление вдоха Истока",
    "Древний Ветер"
]

for col in num_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".")   # если вдруг есть запятые
        .astype(float)
    )
```

```python
#удалим столбец с индексами

df.drop(columns=["Номер пометки"], inplace=True)
```
